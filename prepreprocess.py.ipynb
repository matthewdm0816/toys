```python
# 版本为python3，如果为python2需要在字符串前面加上u
import re
def cut_sent(para):
    para = re.sub('([。！？\?])([^”’])', r"\1\n\2", para)  # 单字符断句符
    para = re.sub('(\.{6})([^”’])', r"\1\n\2", para)  # 英文省略号
    para = re.sub('(\…{2})([^”’])', r"\1\n\2", para)  # 中文省略号
    para = re.sub('([。！？\?][”’])([^，。！？\?])', r'\1\n\2', para)
    # 如果双引号前有终止符，那么双引号才是句子的终点，把分句符\n放到双引号后，注意前面的几句都小心保留了双引号
    para = para.rstrip()  # 段尾如果有多余的\n就去掉它
    # 很多规则中会考虑分号;，但是这里我把它忽略不计，破折号、英文双引号等同样忽略，需要的再做些简单调整即可。
    return para.split("\n")
```

```python
with open("神们自己.txt", "r", encoding='gb18030') as f:
    text = f.read()
```

```python
from icecream import ic
lines = text.split('\n')
ic(lines[5:10], len(lines))
ic(lines[28])
# Remove chinese tab
for idx, line in enumerate(lines):
    lines[idx] = re.sub('\u3000', '', line)
lines = [line for line in lines if len(line) > 0]
ic(lines[500:500+5], len(lines))
ic(lines[28])
```

```python
ic(cut_sent(lines[30]))
sentences = sum(map(cut_sent, lines), start=[])
ic(sentences[100:105], len(sentences))
```
